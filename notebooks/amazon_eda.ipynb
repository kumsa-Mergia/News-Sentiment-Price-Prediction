# Filter The Data by its Ticker Values

### Store the original DataFrame from the analyzer

```python
original_df = analyzer.df.copy() # Important: create a copy so you can restore it
```

### Filter the data and assign it to 'amazon_df'

```python
amazon_df = original_df[original_df["stock"] == "AMZN"].copy()
print("amazon.head()")
print(amazon_df.head())
print("\n")
```

## --- TEMPORARILY SET THE ANALYZER'S DATAFRAME TO THE FILTERED DATA ---

```python
# This is the crucial step to make your existing methods work
analyzer.df = amazon_df
```

### run all analysis methods directly on the 'analyzer' object

### They will now operate on the 'amazon_df' because analyzer.df points to it

```python
print("\n--- Descriptive Statistics for Headline lengths ---")
analyzer.textual_lengths() # Now this will work on the filtered data
analyzer.descriptive_statistics()

print("\n--- Headline Length Visualization ---")
analyzer.headline_length_visualization()

print("\n--- Articles per Publisher Analysis ---")
analyzer.articles_per_publisher()
analyzer.top_publishers()
analyzer.publisher_visualization()

print("\n--- Date-Based Analyses ---")
analyzer.prepare_dates() # This will prepare dates for the filtered data
analyzer.articles_by_day_of_week()
analyzer.articles_over_time(resampling_freq='D')
analyzer.articles_by_month()
analyzer.articles_by_hour()
```
